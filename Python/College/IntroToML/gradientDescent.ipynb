```python
import numpy as np
import matplotlib.pyplot as plt
```

## Initializaion of Variables & Dataframes

```python
x = np.array([0.8,1,1.2,1.4,1.6,1.8,2,2.2,2.4,2.6])
y = np.array([0.7,0.65,0.9,0.95,1.1,1.15,1.2,1.4,1.55,1.5])

m = float(len(x))
print(x.shape)
print(y.shape)
```

```python
theta0 = 0.35
theta1 = 0.95
alpha = 0.01
epoch = 1000
cost_list = []

for i in range(epoch):

    h = theta0 + theta1*x
    djdtheta0 = (1/m)*sum(h-y)
    djdtheta1 = (1/m)*sum((h-y)*x)
    cost = (1/(2*m))*(sum(h-y)**2)

    cost_list.append(cost)

    theta0 = theta0 - alpha*djdtheta0
    theta1 = theta1 - alpha*djdtheta1

print(f"theta0: {theta0}")
print(f"theta1: {theta1}")
#print(cost_list)

np.asarray(cost_list)
```

```python
plt.plot(cost_list)
```
